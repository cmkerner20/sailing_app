```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob

from sklearn.preprocessing import StandardScaler
from torchtext import data

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

```

```python

# this needs to be cleaned up and data can be grouped in a more efficient way 

path= '/Users/tancredicp/Desktop/sailing_app/data/ML Training Resources/'
all_files = glob.glob(path + "/*.csv")


Data = []
for filename in all_files:
    
    
    
    
    # how many point from the start you have to throw away ?
    roll_param2 = 10
    roll_param1 = 3
    offset =  1
    

    dfSpecific = pd.read_csv(filename, index_col=None, header=0)
    dfSpecific.fillna(0, inplace= True)
    dfSpecific['time'] = pd.to_datetime(dfSpecific['time']).dt.tz_localize(None)
    
    
    sogDelta = dfSpecific['sog'].diff() / dfSpecific['time'].diff().dt.total_seconds()
    cogDelta = dfSpecific['sog'].diff() / dfSpecific['time'].diff().dt.total_seconds()
    
    dfFeatures = dfSpecific[['sog', 'time']]
    
    #YOU NEED TO CHANGE THIS..
#     dfFeatures.set_index('time')
    
    dfFeatures['sogDelta'] = sogDelta
    dfFeatures['sogDeltaRoll1'] = dfFeatures[['sogDelta']].rolling(roll_param1, center = True).mean()
    dfFeatures['sogDeltaRoll1'].fillna(dfFeatures['sogDelta'], inplace=True)
    dfFeatures['sogDeltaRoll2'] = dfFeatures[['sogDelta']].rolling(roll_param2, center = True).mean()
    dfFeatures['sogDeltaRoll2'].fillna(dfFeatures['sogDeltaRoll1'], inplace=True)
    
    
    dfFeatures['cogDelta'] = cogDelta
    dfFeatures['cogDeltaRoll1'] = dfFeatures[['cogDelta']].rolling(roll_param1, center = True).mean()
    dfFeatures['cogDeltaRoll1'].fillna(dfFeatures['cogDelta'], inplace=True)
    dfFeatures['cogDeltaRoll2'] = dfFeatures[['cogDelta']].rolling(roll_param2, center = True).mean()
    dfFeatures['cogDeltaRoll2'].fillna(dfFeatures['cogDeltaRoll1'], inplace=True)
    
    scaler = StandardScaler()
    arrayFeaturesScaled = scaler.fit_transform(dfFeatures)
    
    
    
    dfFeatures = pd.DataFrame(arrayFeaturesScaled, columns=dfFeatures.columns)
    
    #tags
    dfTag = dfSpecific[['tag']]
    dfTag.fillna(0, inplace= True)
    
    TAGS = data.Field(unk_token = None)
    TAGS.build_vocab(['0','A','B','C','D','G','T','S'])
    dfTag['tag'] = np.array(TAGS.numericalize((np.array(dfTag['tag']))))[0]
    
    dfFeatures['tag'] =  dfTag['tag']
    
    #drop first value since there will be nans for delta fields
    dfFeatures = dfFeatures[1:]
    Data.append(dfFeatures)
```

```python
MasterFrame = pd.concat(Data)
```

```python
fig, ax = plt.subplots(figsize = (50,10))
ax.plot(MasterFrame.index[:800], MasterFrame.cogDeltaRoll1[:800])
ax.plot(MasterFrame.index[:800], MasterFrame.cogDeltaRoll2[:800])
ax.plot(MasterFrame.index[:800], MasterFrame.cogDelta[:800])
```

```python
MasterFrame.columns
```

```python
X = MasterFrame[['sog', 'sogDelta', 'sogDeltaRoll', 'cogDelta', 'cogDeltaRoll']]
y = MasterFrame[['tag']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 3, test_size = 0.8, shuffle = False)
```

```python
RFC = RandomForestClassifier(n_estimators= 100, max_depth =  4)
RFC.fit(X_train,y_train)
```

```python
RFC.score(X_train, y_train), RFC.score(X_test, y_test)
```
